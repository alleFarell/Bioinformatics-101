```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

# Searching Target Protein

### Target search for Coronovirus

```python
# Target search for Coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### Select and retrieve bioactivity data for *SARS coronavirus 3C-like proteinase*

```python
selected_protein_id = targets.target_chembl_id[4]
selected_protein_id
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_protein_id).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
pd.options.display.max_columns = None
df.head()
```

Save the resulting bioactivity data to csv file

```python
df.to_csv('coronavirus_01_bioactivity_data_raw.csv', index=False)
```

# Preprocessing Bioactivity Data

### Missing Values Handling

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df2.canonical_smiles.notna()]
df2
```

### Drop Duplicate Data in Canocical SMILES

```python
df2.duplicated(['canonical_smiles']).sum()
```

```python
df2.drop_duplicates(['canonical_smiles'], inplace=True)
df2.shape
```

### Labeling Compunds Threshold as Active, Intermadiate, or Inactive
The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**. 

```python
bioactivity_threshold = []
for i in df2['standard_value']:
    if float(i) < 1000:
        bioactivity_threshold.append('Active')
    elif float(i) > 10000:
        bioactivity_threshold.append('Inactive')
    else:
        bioactivity_threshold.append('Intermediate')
```

Converting bioactivity threshold to pandas Series

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
bioactivity_class
```

### Combine 3 important columns + bioactivity class into Dataframe

```python
select = ['molecule_chembl_id','canonical_smiles','standard_value']
df_fix = pd.concat([df2[select].reset_index(), bioactivity_class], axis=1)
df_fix
```

```python
df_fix.to_csv('coronavirus_02_bioactivity_data_curated.csv', index=False)
```
